# **Encoding Dataset**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
df['encoded_material'] = encoder.fit_transform(df['material'])
```

# **Splitting Data**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor


# Separate the features (X) and the target element columns (y)
X = df.drop(columns=['material', 'encoded_material']).values
y = df[df.columns[:df.columns.get_loc('critical_temp')]].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

# **Models**

##linear regression


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.multioutput import MultiOutputRegressor
```

```python
# Create a Linear Regression as the base model
base_model = LinearRegression()
```

```python
# Create a MultiOutputRegressor with the base model
model = MultiOutputRegressor(base_model)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python

# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Optionally, if you want to predict the material strings for each sample:
predicted_materials = []
for elem_values in y_pred:
    material = "".join([str(int(round(elem))) for elem in elem_values])
    predicted_materials.append(material)
```

```python
# Print the predicted materials for the test set
for temp, material in zip(X_test[:, -1], predicted_materials):
    print(f"Critical Temperature: {temp}, Predicted Material: {material}")
```

```python
def find_closest_temperature_with_range(temperature_list, target_temperature, min_temp, max_temp):
    if min_temp <= target_temperature <= max_temp:
        closest_temperature = None
        min_difference = float('inf')

        for temperature in temperature_list:
            difference = abs(temperature - target_temperature)
            if difference < min_difference:
                min_difference = difference
                closest_temperature = temperature

        return closest_temperature
    else:
        return target_temperature

#find_closest_temperature_with_range(temperatures, target_temp, min_temp, max_temp)
```
